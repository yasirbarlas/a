```python
#Importing several modules required to run code

from cryptocmd import CmcScraper
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt
```

```python
#coin = input("Input the coin(code) you want to analyse:")
coin = "XRP"

scraper = CmcScraper(coin)
df = scraper.get_dataframe()

close = df["Close"]
dates = df["Date"]

print("Here are the first and last 5 dates since the coin was launched:")
print("\n")
print(df)
x, y = df.shape
print("\n")
print("The coin launched", x, "days ago.")

print("\n")
print("Here is a plot of how the coin's value has changed over time:")
plt.figure(figsize=(12, 6), dpi=80)
plt.plot(dates, close)
plt.title("Plot of coin overtime")
plt.xlabel("Date")
plt.ylabel("Value per coin")
plt.show()
```

```python
#user_date = input("Input a date(DD-MM-YYYY) to see the difference in value from then to now:")
user_date = "20-05-2021"

today = dt.date.today()
yesterday = today - dt.timedelta(days=1)

def difference(df, start):
    a = df[df["Date"] == start]
    b = df[df["Date"] == str(yesterday)]
    print("Here is the coin information of yesterday and the date you have chosen:")
    print("\n")
    print(a)
    print(b)
    print("\n")
    c = b.iloc[:,4].values - a.iloc[:,4].values
    c = round(c[0], 3)
    if c >= 0:
        print("You have not made a loss, the profit you would have made per coin is", c)
    elif c < 0:
        print("You have made a loss, the loss you would have made per coin is $", c)
    else:
        print("Ensure that you are using the correct date format, or the coin did not exist during this period")
    return ""
    
final = difference(df, user_date)
print(final)
```
